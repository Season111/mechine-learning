```python
#求矩阵的转置，矩阵的逆，矩阵的奇异值分解，计算线性方程[[2,8], [1,5]]*[[x],[y]]=[[9],[31]]

import numpy as np

b = np.arange(24).reshape(2,3,4) #改变数组形状

b.ravel() #将多维数组变成一维数组

b.flatten() # 与ravel()相同，但是flatten()返回的是真实的数组，需要分配新的内存空间，而ravel()仅仅是改变视图。

b.shape = (6,4) #改变数组形状

b.transpose() #转置



a = np.arange(9).reshape(3,3)

b = np.arange(9).reshape(3,3)

np.hstack((a,b))  # 水平叠加注意 这里是两层括号

np.vstack((a,b)) #垂直叠加

np.dstack((a,b)) #深度叠加

np.hsplit(a,3) # 横向拆分

np.vsplit(a,3) # 纵向拆分

c= np.arange(27).reshape(3,3,3)

np.dsplit(c,3)  #深度拆分

#计算线性方程[[2,8], [1,5]]*[[x],[y]]=[[9],[31]]
a=np.array([[2,8],[1,5]])
b=np.array([[9],[31]])
np.linalg.solve(a,b)
print(np.linalg.solve(a,b))
```
