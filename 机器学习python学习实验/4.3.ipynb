```python
#补充类中的push和pop函数，需要注意异常处理
class Stack():

    def __init__(self, size=5):

        self.size = size

        self.stack = []

        self.top = -1

    def is_full(self):

        return self.top+1 == self.size

    def is_empty(self):

        return self.top == -1

    def __repr__(self):

        return str(self.stack)
    def push(self, value):

        if self.is_full():

            raise Exception('Stack is full')

        self.stack.append(value)

        self.top += 1
    def pop(self):

        if self.is_empty():

            raise Exception('Stack is empty')

        self.top -= 1

        return self.stack.pop()
```

```python
class Queue():

    def __init__(self, size=5):

        self.size = size

        self.front = 0

        self.rear = 0

        self.queue = []

    def is_full(self):

        return self.rear - self.front == self.size

    def is_empty(self):

        return self.front == self.rear

    def __repr__(self):

        return str(self.queue)
    
    def in_Queue(self, value):

        if self.is_full():

            raise Exception('Queue is full')

        self.queue.append(value)

        self.rear += 1
    
    def out_Queue(self):

        if self.is_empty():

            raise Exception('Queue is empty')

        self.front += 1

        return self.queue.pop(0)
```
